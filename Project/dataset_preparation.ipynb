## Feature Engineering

```python
df_clean['Inception'] = pd.to_datetime(df_clean['Inception'])
```

```python
df_clean.head()
```

```python
# Cleaning Expense Ration Column
exp_ratio = []
exp = df_clean['ExpenseRatio']
exp = list(exp)
for i in exp:
    exp_ratio.append(float(i.split(' ')[0]))
df_clean['ExpenseRatio'] = exp_ratio
```

```python
df_clean.head()
```

```python
# Cleaning Turnover Ration Column
turn_ratio = []
turn = df_clean['TurnoverRatio']
turn = list(turn)
for i in turn:
    turn_ratio.append(float(i.split('(')[0]))
df_clean['TurnoverRatio'] = turn_ratio
df_clean.head()
```

```python
df_clean.info()
```

```python
## choosing scheme which are 10 yrs old or more
df_clean = df_clean[df_clean['Inception'] <= '2015-03-31']
df_clean
```

```python
cols = df_clean.columns[6:]
for i in cols:
    print(df_clean[df_clean[i] == '–'][i])
```

```python
indx = df_clean[(df_clean['ALP10Y']=='–') | (df_clean['BET10Y']=='–') | (df_clean['SD_10YRS']=='–') | (df_clean['R2_10YRS']=='–') | (df_clean['SHP_10YRS']=='–')].index
df_clean.drop(indx,inplace=True)
```

```python
## formatting the other performance and risk ratios
cols = df_clean.columns[6:]
for i in cols:
    df_clean[i] = df_clean[i].astype(float)
df_clean.info()
```

```python
df_clean['SchemeCode'] = df_clean['SchemeCode'].astype(int)
```

```python
df_clean.info()
```

```python
df_clean = df_clean.reset_index(drop=True)
df_clean
```

```python
df_clean.to_csv('fodder2.csv',index = False)
```

## Debt-Oriented Funds
These are primarily suited for conservative investors seeking steady returns with lower risk.

1. Low Duration (130)

Invest in debt & money market instruments with a duration of 6 to 12 months.

Lower interest rate risk, good for short-term parking of funds.

2. Short Duration (77)

Duration: 1–3 years.

Balanced risk-return profile; better returns than liquid/low duration funds.

3. Money Market (64)

Invest in money market instruments (T-bills, CPs, CDs).

Tenure usually up to 1 year. Suitable for very short-term needs.

4. Ultra Short Duration (51)

Invest in instruments with 3–6 months duration.

Slightly higher returns than liquid funds with minimal volatility.

5. Medium Duration (39)

Duration: 3–4 years.

Moderate risk; impacted by interest rate changes.

6. Medium to Long Duration (39)

Duration: 4–7 years.

Higher interest rate risk; better during falling rate scenarios.

7. Corporate Bond (34)

Invest 80%+ in highest-rated corporate bonds.

Lower risk, stable returns.

8. Banking & PSU (41)

Invest in debt instruments of banks, PSUs, PFIs.

Good credit quality and safety.

9. 10 yr Government Bond (12)

Invest in government securities with 10-year maturity.

High interest rate sensitivity; suitable during rate cuts.

## Equity-Oriented Funds
Suitable for long-term investors seeking capital appreciation.

1. Large-Cap (82)

80%+ in top 100 companies by market cap.

Relatively stable, lower volatility.

2. Large & Mid-Cap (68)

Minimum 35% in both large- and mid-cap stocks.

Diversified exposure with growth potential.

3. Mid-Cap (57)

65%+ in mid-cap stocks (101–250th by market cap).

Higher risk and return potential.

4. Small-Cap (30)

65%+ in small-cap stocks (251+ ranked).

High risk, high return, volatile.

5. Multi-Cap (17)

25% each in large-, mid-, and small-cap.

Well-diversified across market caps.

6. Value (34)

Focus on undervalued stocks with turnaround potential.

Contrarian strategy; longer investment horizon needed.

7. Contra (12)

Invest against market trends.

Contrarian approach, requires patience and timing.
